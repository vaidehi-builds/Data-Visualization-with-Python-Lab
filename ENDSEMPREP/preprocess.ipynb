Data Preprocessing and Cleaning

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("FileName")
miss=df.isna().sum()
print(miss)
```

```python
df.replace([-1,999,'NA','N/A'],np.nan,inplace=True)
print("Missing values after treatment")
print(df.isna().sum())
```

```python
df.fillna(df.mean(),inplace=True)
df.fillna(method='ffil', inplace=True)
df.fillna(method='bfill',inplace=True)
df.dropna(axis=1,thresh=(0.7*len(df)),inplace=True)

print("Missing values after treatment")
print(df.isna().sum())
```

```python
column='Salary'
Q1=df[column].quantile(0.25)
Q3=df[column].quantile(0.75)
IQR=Q3-Q1
upper=Q1-1.5*IQR
lower=Q3+1.5*IQR
outliers=df[df[column>upper]|df[column]<lower]
print("outliers: ")
print(outliers)
med=df[column].median()
df.loc[df[column]>upper|df[column]<lower]=med
print("After handling outliers: ")
print(df)
```

```python
#handling duplicates
df_dup=df.duplicated() #returns True/False
print("Duplicated ROws: ")
print(df_dup)

print("Duplicated Records")
print(df[df_dup])

df=df.drop_duplicates()
```
