```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#IRIS DATASET
df_raw=pd.read_csv("IRIS.csv")
df=df_raw.copy()
print("First 5 rows:")
print(df.head())
print("Basic Structure")
print(df.info())


```

```python
col=['sepal_length','petal_length','petal_width']
print("SUmmary Statistics: ")
print(df[col].agg(['mean','median','min','max']).round(3))
```

```python
# BOXPLOT: 
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
df.boxplot(column='sepal_width',by='species',ax=plt.gca())
plt.title("Relationship btw Species and sepal width")
plt.suptitle("")
plt.xlabel("FLower Species")
plt.ylabel("Sepal Width")
# plt.show() only for independent graphs ( in suplot not needed, it will break the graph)

plt.subplot(1,2,2)
sns.boxplot(x='species',y='sepal_width', data=df)
plt.title("Distribution of Sepal WIdth over different Categories")

plt.tight_layout()
plt.show()
```

```python
# Scatter
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.scatter(df['petal_length'],df['petal_width'],s=10)
plt.title("Matplotlib Scatter")
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.subplot(1,2,2)
sns.scatterplot(x='petal_length',y='petal_width',data=df,s=18,hue='species',style='species')
plt.title("SEABORN")
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
df.boxplot(column='petal_length',by='species',ax=plt.gca())
plt.title("PANDAS")
plt.suptitle("")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.subplot(1,2,2)
sns.boxplot(data=df,x='species',y='petal_length')
plt.title("SEABORN")
plt.xlabel("Species")
plt.ylabel("Petal Length")
plt.tight_layout()
plt.show()
```
