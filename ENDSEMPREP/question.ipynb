```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("SampleSuperstore.csv")
```

```python
cat_pr=df.groupby('Category')['Profit'].sum()
plt.figure(figsize=(8,6))
plt.bar(cat_pr.index,cat_pr.values)
plt.title("Total Sales Profit in different Categories")
plt.xlabel("Categories")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```

```python
subcat_count=df['Sub Category'].value_counts()
plt.figure(figsize=(8,6))
plt.bar(subcat_count.index,subcat_count.values,color='skyblue')
plt.title("Number of transactions per Sub-Category")
plt.xlabel("Sub Category")
plt.ylabel("Number of Transactions")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(df['Discount'],df['Profit'])
plt.title("Relationship between Discount and Profit")
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.histplot(x='Profit',data=df,bins=20)
plt.title("Distribution of Profit")
plt.xlabel("Profit")
plt.tight_layout()
plt.show()
```
