```python
def test(a,b):
    if(a==b):
        ans=a+b
    else:
        ans=a*b
    return ans
print(test(3,3))
print(test(3,4))
```

```python
def long(text):
    words=text.split()
    word_len=[len(word)for word in words]
    longest=words[word_len.index(max(word_len))]
    print(longest)
long("Hi I am you biggest supporter")
```
